```python
def plot_graph(acc_record, fileName, isTrain, error = False):
    if error:  
        acc_record = [1-tmp for tmp in acc_record]
        if isTrain:
            yLabel = 'Train error'
        else:
            yLabel = 'Test error'
    else:
        if isTrain:
            yLabel = 'Train accuracy'
        else:
            yLabel = 'Test accuracy'
    plt.figure(1)
    plt.plot(range(len(acc_record)), acc_record)
    plt.xlabel(str(len(acc_record)) + ' iterations')
    plt.ylabel(yLabel)
    plt.savefig(os.path.join(GRAPH_DIR, fileName))
    plt.show()
```
