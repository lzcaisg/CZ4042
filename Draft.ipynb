```python
def scatter_plot(x, y, save_name, batch_size):
    plt.figure(1)
    plt.scatter(x=x, y=y, s=2)
    plt.xlabel('Batch Size: ' + batch_size)
    plt.ylabel('Time Needed')
    plt.savefig(save_name)
    plt.show()
    
def box_plot(x, batch_size):
    plt.figure(1)
    plt.boxplot(x = x, showfliers=False)
    plt.xlabel('Batch Size: ' + str(BATCH_SIZE))
    plt.ylabel('Time Needed')
    plt.savefig(save_name_box)
    plt.show()
```

```python
file_name = "Q1_Result/Output_BatchSize/PartA-Time_Usage-"+str(EPOCHS)+"-4.out"

with open(file_name, 'rb') as handle:
    b = pickle.load(handle)
    
batch_size_list = [64, 32, 16, 8, 4]
for i in range(len(batch_size_list)):
    BATCH_SIZE = batch_size_list[i]
    save_name_scatter = "Q1_Result/Plot_BatchSize/PartA-Epoch_Time-"+str(EPOCHS)+'-'+str(BATCH_SIZE)+".png"
    save_name_box = "Q1_Result/Plot_BatchSize/PartA-Epoch_Time-BOX-"+str(EPOCHS)+'-'+str(BATCH_SIZE)+".png"
    
    time_array = b[i]
    print("Average training time for batch_size", BATCH_SIZE, "is:", sum(time_array)/len(time_array))
    
    scatter_plot(x=range(EPOCHS), y=time_array, save_name=save_name_scatter, batch_size=str(BATCH_SIZE))
    box_plot(x = time_array, batch_size = str(BATCH_SIZE))
    
```
